# Customer Segmentation Using K-Means Clustering

![K-Means Clustering](Images/customer-segmentation.png)

This project focuses on customer segmentation using the K-Means clustering algorithm. By analyzing a dataset of mall customers, we aim to identify distinct groups based on their demographics and spending behaviors. The project includes the following steps:

* Data Preparation: Load and preprocess the dataset to make it suitable for clustering.
* Exploratory Data Analysis (EDA): Visualize key features to understand patterns and relationships in the data.
* Feature Engineering: Normalize data and select relevant features for clustering.
* Clustering: Apply the K-Means algorithm and determine the optimal number of clusters using the Elbow Method.
* Visualization: Use PCA to reduce dimensionality and visualize the identified clusters.
* Evaluation: Assess the quality of clusters using metrics like Silhouette Score and Davies-Bouldin Index.
  
The results provide insights into customer behaviors, which can help businesses tailor their marketing strategies and improve customer experiences.
___
## Types Clustering:

### Connectivity-based clustering
* Distance based
* E.g., Hierarchical clustering


### Centroid-based clustering
* Represents each cluster by a single mean vector
* E.g., k-means algoritm


### Distribution-based clustering
* Modeled using statistical distributions
* E.g., Multivariate normal distributions used by the expectation-maximization algorithm.


### Density-based clustering
* Defines clusters as connected dense regions in the data space.
* E.g., DBSCAN

# IV: K-Means Cluster Creation

![K-Means Cluster Creation](Images/clustering-process.png)

In this step, we perform the following:
1. Select relevant features for clustering.
2. Normalize the data to ensure features contribute equally.
3. Use the **Elbow Method** to determine the optimal number of clusters.
4. Apply **K-Means Clustering** to segment customers.
5. Visualize the clusters in a 2D space using PCA (Principal Component Analysis).
6. Enhance the visualization by adding centroids and cluster boundaries with ellipses.

## Step 1: Select relevant features

```python
# Clustering is based on 'Age', 'Annual Income (k$)', and 'Spending Score (1-100)'
X = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

## Step 2: Normalize the data to ensure all features contribute equally

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Step 3: Determine the optimal number of clusters using the Elbow Method

```python
inertia = []  # To store inertia values for different numbers of clusters
for k in range(1, 11):  # Test 1 to 10 clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
```

## Step 4: Plot the Elbow Curve

```python
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), inertia, marker='o', linestyle='--', color='b')
plt.title('Elbow Method for Optimal Clusters', fontsize=16)
plt.xlabel('Number of Clusters', fontsize=12)
plt.ylabel('Inertia', fontsize=12)
plt.grid(True)
plt.show()
```

## Step 5: Apply K-Means using the optimal number of clusters

```python
# Based on the Elbow Curve, assume 3 clusters are optimal
optimal_clusters = 3
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
kmeans.fit(X_scaled)

# Display the number of iterations the algorithm took to converge
print(f"The algorithm converged after {kmeans.n_iter_} iterations.")
```

## Step 6: Assign cluster labels to the dataset

```python
data['Cluster'] = kmeans.labels_
```

## Step 7: Visualize clusters with PCA

```python
# Reduce dimensions to 2 for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Transform centroids into PCA space
centroids_pca = pca.transform(kmeans.cluster_centers_)

# Scatter plot of PCA-transformed data with cluster coloring
sns.scatterplot(
    x=X_pca[:, 0], 
    y=X_pca[:, 1], 
    hue=data['Cluster'], 
    palette='viridis', 
    s=100, 
    alpha=0.7,
    edgecolor='w'
)

# Plot cluster centroids with distinct markers
plt.scatter(
    centroids_pca[:, 0], 
    centroids_pca[:, 1], 
    s=300, 
    c='red', 
    marker='X', 
    label='Centroids'
)

# Draw ellipses around each cluster
for i, centroid in enumerate(centroids_pca):
    # Extract points belonging to the current cluster
    cluster_points = X_pca[data['Cluster'] == i]
    
    # Compute the covariance matrix
    cov = np.cov(cluster_points.T)
    
    # Perform eigen decomposition
    eigenvalues, eigenvectors = np.linalg.eigh(cov)
    
    # Compute width and height of the ellipse
    width, height = 2 * np.sqrt(eigenvalues)
    
    # Compute the angle of rotation
    angle = np.degrees(np.arctan2(*eigenvectors[:, 0][::-1]))
    
    # Add an ellipse to represent the cluster boundary
    ellipse = Ellipse(
        xy=centroid, 
        width=width, 
        height=height, 
        angle=angle, 
        edgecolor='black', 
        facecolor='none', 
        linewidth=2, 
        linestyle='--'
    )
    plt.gca().add_patch(ellipse)

# Final plot decorations
plt.title('Customer Segments with Cluster Centroids (PCA)', fontsize=16)
plt.xlabel('Principal Component 1', fontsize=12)
plt.ylabel('Principal Component 2', fontsize=12)
plt.legend(loc='upper right', title='Clusters')
plt.grid(True)
plt.show()
```

# V. Evaluation

In this step, we evaluate the K-Means clustering model using two metrics:

1. **Silhouette Score**:
   - Measures how well-separated and cohesive the clusters are.
   - Values range from **-1** (poor clustering) to **1** (perfect clustering).
   - A higher score indicates well-defined clusters.

2. **Davies-Bouldin Index**:
   - Evaluates the ratio of within-cluster scatter to between-cluster separation.
   - Values are non-negative, with **lower values indicating better clustering**.

## Metric 1: Silhouette Score

```python
# This measures how similar each point is to its own cluster (cohesion) compared to other clusters (separation).
# Values range from -1 (poor clustering) to 1 (perfect clustering).
silhouette_avg = silhouette_score(X_scaled, kmeans.labels_)
print(f"Silhouette Score: {silhouette_avg:.2f}")
# A higher silhouette score indicates better-defined clusters.
```

## Metric 2: Davies-Bouldin Index

```python
# This evaluates the ratio of within-cluster scatter to between-cluster separation.
# Lower values indicate better clustering.
davies_bouldin = davies_bouldin_score(X_scaled, kmeans.labels_)
print(f"Davies-Bouldin Index: {davies_bouldin:.2f}")
# A lower Davies-Bouldin Index indicates better clustering quality.
```
